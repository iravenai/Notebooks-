```python
# Call Me 
import random

def call_me():
  activations = ["Tanh", "sigmoid",
                 "relu", "Lrelu",
                 "softmax", "binary",
                 "linear"]

  return random.choice(activations)

print(call_me())
```

```python
from sklearn import tree

x = [[0, 0], [1, 1]]
y = [0, 1]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
clf.predict([[2., 2.]]) #Class prediction
clf.predict_proba([[2., 2.]])) #Class proba prediction
```
