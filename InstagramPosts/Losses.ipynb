```python
# Mean Squared Error
def mse( target, prediction ) :
  return  np.sum((target - prediction) ** 2) / np.size(target)
```

```python
# Mean Absolute Error
def mae( target, prediction ) :
  return np.sum( np.abs( target - prediction ) ) / np.size( target )
```

```python
# Huber Loss
def Huber( target, prediction, delta ) :     
  condition = np.abs( target - prediction ) < delta
  l =  np.where( condition, 0.5 * ( target - prediction ) **2, 
                delta * ( np.abs( target - prediction ) - 0.5 * delta ) )
  return np.sum( l ) / np.size( target )
```

```python
# Binary Loss
def cross_entropy( target, prediction ) :
  return - np.sum( target * np.log( prediction ) + 
                  (1 - target ) * np.log(1 - prediction ))/
                  np.size(target)
```

```python
# Hinge Loss
def hinge( target, prediction ) :
  l = 0
  size = np.size( prediction )
  for i in range( size ) :
    l = l + max( 0, 1 - target[i] * prediction[i] )
  return l / size
```

```python
#KL Divergence
def KL(a, b):
    a = np.asarray(a, dtype=np.float)
    b = np.asarray(b, dtype=np.float)
    return np.sum(np.where(a != 0, a * np.log(a / b), 0))
```
