```python
#_________________________________________________________________________________________________________________________#
#______________________________________________________Raven.ai___________________________________________________________#
#_________________________________________________________________________________________________________________________#

# L'ets tackle the XOR gate problem, with the simpliest neural network.

# The XOR gate : 0 XOR 0 --> 0
#                1 XOR 0 --> 1
#                0 XOR 1 --> 1
#                1 XOR 0 --> 0    

# In oder th achieve this we need to build a neural net with one hidden layer to compute the input value on two steps

# Architecture : Input Layer  (2 features)
#                Hidden Layer (3 features)    
#                Output Layer (1 features)
```

```python
import tensorflow as tf
import numpy as np
```

```python
# L'ets create our data with all possible input and output values for our XOR

x = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [1],
              [1],
              [0]])

# we need to convert the data to tensor in order to feed our neural net

x = tf.convert_to_tensor(x, dtype=tf.float16)
y = tf.convert_to_tensor(y, dtype=tf.float16)
```

```python
# L'ets initialize our matrices weights, we will do it randomly and then let the network adjust it self

w1 = tf.Variable(np.random.randn(2, 3), dtype=tf.float16)
w2 = tf.Variable(np.random.randn(3, 1), dtype=tf.float16)
```

```python
# Now we need a function to propagte x throught the network
@tf.function
def forward(x, w1, w2):
  x = tf.sigmoid(tf.matmul(x, w1))
  x = tf.sigmoid(tf.matmul(x, w2))
  return x
```

```python
# printing the result and see what next
print(forward(x, w1, w2))
```

```python
# The network is returning randoms values, we need to implement a loss function in order to tel the NN how far is it from the right target value

import tensorflow.keras.backend as k

@tf.function
def loss(predicted_y, target_y):
  return tf.reduce_mean(k.binary_crossentropy(target_y, predicted_y))
```

```python
# now we got our data, propagate func, and a loss func we still nedd a function to train our NN 

@tf.function
def train(inputs, outputs, learning_rate):
  with tf.GradientTape() as tape:
    current_loss = loss(forward(x, w1, w2), outputs)
  #once done we have to calculate the gradient between loss and matrix weight
  dw1, dw2 = tape.gradient(current_loss,[w1,w2])
  #after that we need to adjust the weights by back propagation 
  w1.assign_sub(learning_rate * dw1)
  w2.assign_sub(learning_rate * dw2)
  del tape
```

```python
# l'ets define some parameters for our trainnig loop and lunch the training

epochs = range(10000)
learning_rate = 0.1

for epoch in epochs:
  current_loss = loss(forward(x,w1,w2), y)
  train(x, y, learning_rate)
  print("Current loss: {:2.3f}".format(current_loss));
```

```python
#Testing our model

x_ = tf.convert_to_tensor(np.array([[1, 1]]), dtype=tf.float16)
print(forward(x_, w1, w2).numpy())

x_ = tf.convert_to_tensor(np.array([[1, 0]]), dtype=tf.float16)
print(forward(x_, w1, w2).numpy())
```
