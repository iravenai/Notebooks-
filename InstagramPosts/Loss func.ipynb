```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Binary Step
# returns 0 if x < 0, 1 in the other case
def binary_step(x):
  return np.heaviside(x,1)

fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, binary_step(x), color="#7f64b0")
plt.title("binary step")
plt.show()
```

```python
# Linear
# returns x as x
def linear(x):
  return x

fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, linear(x), color="#7f64b0")
plt.title("linear")
plt.show()
```

```python
# Sigmoid
# must know !!! 1 / (1 + exp(-x))
def sigmoid(x):
  return 1/(1+np.exp(-x))


fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, sigmoid(x), color="#7f64b0")
plt.title("sigmoid")
plt.show()
```

```python
# TanH
# must know : (1-exp(-2x)) / (1+exp(-2x))

def tanh(x):
  return (1-np.exp(-2*x))/(1+np.exp(-2*x))

fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, tanh(x), color="#7f64b0")
plt.title("tanh")
plt.show()
```

```python
# Relu : the sexiest one
# return 0 if x < 0, x if x >0

def relu(x):
  x1 = []
  for i in x :
    if i < 0:
      x1.append(0)
    else:
      x1.append(i)
  return x1

fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, relu(x), color="#7f64b0")
plt.title("relu")
plt.show()
```

```python
# Softmax:
# return the softmax value exp(x)/sum(exp(x))

def softmax(x):
  return np.exp(x)/ np.sum(np.exp(x), axis= 0)

fig = plt.figure()
fig.patch.set_facecolor("#f8f8f8ff")
x = np.linspace(-10, 10)
ax = plt.axes()
ax.set_facecolor("#f8f8f8ff")
plt.plot(x, softmax(x), color="#7f64b0")
plt.title("softmax")
plt.show()
```
